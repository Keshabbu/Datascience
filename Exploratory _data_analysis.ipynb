```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df  =pd.read_csv(url, header=None,na_values='?')
```

```python
df.columns=["age", "workclass", "fnlwgt", "education",
            "education_num","marital_status", "occupation",
            "relationship", "race", "sex", "capital_gain",
            "capital_loss", "hours_per_week", "native_country", "income"]
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['age'],bins=20,kde = True)
plt.xlabel('age')
plt.ylabel('frequancy')
plt.title('distribution of age')
plt.show()
```

```python
print(f'avrage working per week {df['hours_per_week'].mean().round(2)}')
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['hours_per_week'],bins=20,kde = True)
plt.xlabel('hours_per_week')
plt.ylabel('frequancy')
plt.title('distribution of working hour per week')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(df,y='education',order = df['education'].value_counts().index)
plt.xlabel('education')
plt.ylabel('frequancy')
plt.title('distribution of education')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(df,y='income',order = df['income'].value_counts().index)
plt.xlabel('income')
plt.ylabel('frequancy')
plt.title('distribution of income')
plt.show()
```

```python
print(df['income'].value_counts(normalize= True))
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df,x='income',y='age',order = df['income'].value_counts().index)
plt.xlabel('income')
plt.ylabel('age')
plt.title('boxplot of income and age')
plt.show()
```

```python
print(df['income'].value_counts(normalize= True))
```

```python
edu_group = df.groupby('education') ['income'].mean().sort_values(ascending = True).head(10)
```
